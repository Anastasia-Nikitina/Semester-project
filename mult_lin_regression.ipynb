```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as lng

#масштабируем значения
data = pd.read_csv("./50_Startups.csv")
x1 = (data['R&D Spend'])/100000
x2 = (data['Marketing Spend'])/100000
x0 =  x1/x1
y = (data['Profit'])/100000

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x1, x2, y, c = 'green')
ax.set_xlabel('R&D Spend')
ax.set_ylabel('Marketing Spend')
ax.set_zlabel('Profit')


X = pd.concat ([x0, x1, x2], axis = 1)
X_T = pd.DataFrame.transpose(X)
theta = ((lng.inv(X_T.dot(X))).dot(X_T)).dot(y)

ax.plot([min(x1), max(x1)], [min(x2), max(x2)], [theta[0] + min(x1)*theta[1] + min(x2)*theta[2], theta[0] + max(x1)*theta[1] + max(x2)*theta[2]], c='purple')
plt.show()
```
